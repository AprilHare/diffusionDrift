```python
import basicSim
import pandas as pd
import numpy as np
import json as json

params = {'vx':0, 'vy':0, 'D':1, 't':20, 'N':50}
trajectory = basicSim.runSim( [ params['vx'], params['vy'] ], abs( params['D'] ), max( int( params['t'] ), 1), max( int( params['N'] ), 1) )

sim_data = []
for i in range(params['t']):
    x= [i for j in range(len(trajectory[:,:,i].tolist()))]
    sim_data.append(zip(x,trajectory[:,0,i].tolist(),trajectory[:,1,i].tolist()))
x = np.asarray(sim_data).tolist()

[j[0] for j in x for j[0] in j]

#[[{"t": j[0], "x":j[1], "y": j[2]} for j in i for k in x]



```

```python
ndata = 100
x = 10 * np.random.rand(ndata) - 5
y = 0.5 * x + 0.5 * np.random.randn(ndata)
A = 10. ** np.random.rand(ndata)
c = np.random.rand(ndata)
json.dumps([{"_id": i, "x": x[i], "y": y[i], "area": A[i],"color": c[i]} for i in range(ndata)])
```

```python
import StringIO

f = StringIO.StringIO()
#f = open('static/data.tsv', 'w')
f.write("time\tx\ty\n")
for i in sim_data:
   for j in i:
    f.write(str(j[0])+"\t"+str(j[1])+"\t"+str(j[2])+"\n")
tsv = f

f.close
```
